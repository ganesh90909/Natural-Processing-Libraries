# Bigrams

```python
cocoa_words=reuters.words(categories='cocoa')
len(cocoa_words)
```

```python
cocoa_words_condensed=cocoa_words[:100]
cocoa_words_condensed
```

```python
from nltk.corpus import stopwords

cocoa_words_condensed=[w.lower() for w in cocoa_words_condensed if w.lower() not in stopwords.words('english')]
cocoa_words_condensed[:10]
```

```python
#removing punctuation
import string

cocoa_words_condensed=[w for w in cocoa_words_condensed if w not in string.punctuation]
len(cocoa_words_condensed)
```

```python
#unigrams
from nltk import ngrams
cocoa_unigrams=list(ngrams(cocoa_words_condensed,1))
cocoa_unigrams
```

```python
#bigrams
cocoa_bigrams=list(ngrams(cocoa_words_condensed,2))
cocoa_bigrams
```

```python
from nltk import FreqDist
freq_unigrams=FreqDist(cocoa_unigrams)
freq_unigrams
```

```python
freq_bigrams=FreqDist(cocoa_bigrams)
freq_bigrams
```

```python
import matplotlib.pyplot as plt

freq_bigrams.plot(5,cumulative=False)
```

# Stemming

```python
from nltk.stem import PorterStemmer,LancasterStemmer,SnowballStemmer
```

```python
porter=PorterStemmer()
lancaster=LancasterStemmer()
snowball=SnowballStemmer('english')

print(porter.stem('lightening'), lancaster.stem('lightening'),snowball.stem('lightening'))
print(porter.stem('surviving'), lancaster.stem('surviving'),snowball.stem('surviving'))
```

# Lemmatization

Lemmatization aims to achieve genuine dictionary root word but not a truncated ersion of a word like stemming

```python
from nltk import WordNetLemmatizer
```

```python
wnl=WordNetLemmatizer()
wnl.lemmatize('surviving')
```

```python
wnl.lemmatize('lightening')
```

```python
#we can change the parts of speech of lemmatization. The default pos os Noun
wnl.lemmatize('lightening',pos='v')
```
